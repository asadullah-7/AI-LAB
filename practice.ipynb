### Simple BFS traversal

```python
from collections import deque

# Graph
# graph = {
#     'A': ['B', 'C'],
#     'B': ['D', 'E'],
#     'C': ['F'],
#     'D': [],
#     'E': [],
#     'F': []
# }

graph2 = {
    'A':['B', 'C'],
    'B':['G', 'H'],
    'C':['S', 'F'],
    'S':['L'],
    'F':['Q'],
    'Q':['R','S'],
    'G':[],
    'H':[],
    'L':[],
    'R':[],
    'S':[]
}

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()

        if node not in visited:

            print(node, end=" ")
            visited.add(node)

            for neighbor in graph[node]:
                queue.append(neighbor)


bfs(graph2,'A')
```

### BFS with goal challenge

```python
from collections import deque

# Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs_with_goal(graph, start, goal):
    explored = set()
    frontier = deque([(start, [start])])

    while frontier:

        node,path = frontier.popleft()

        if node == goal:
            print("Goal Found!\nPath: ,",path)
            return
        
        if node not in explored:
            explored.add(node)

            # Expand node
            for child in graph[node]:
                if child not in explored:
                    frontier.append((child, path + [child]))

    print("Goal not found")
    

# Run
bfs_with_goal(graph, 'A', 'E')
```

### BFS with cost calculation

```python
graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'F':{'E':0,'G':4},
    'D':{'B':3,'G':8},
    'G':{},
    'E':{'G':2},
    'C':{'F':1}

}


graph2 = {
    'S':{'D':3,'E':9,'P':1},
    'D':{'B':1,'C':8, 'E': 2},
    'E':{'R':1,'H':8},
    'P':{'Q':15},
    'B':{'A':2},
    'C':{'A':3},
    'R':{'F':1},
    'H':{'P':2,'Q':1},
    'F':{'G':2,'C':1},
    'P':{'Q':15},
    'Q':{},
    'A':{},
    'G':{}

}


def bfs_with_cost(graph,start,goal):

    explored = set()
    frontier = deque([(start, [start],0)])

    while frontier:
        node,path,cost = frontier.popleft()

        if node == goal:
            print("Goal Found! \nPath, ",path, " And cost is: ",cost)
            return

        if node not in explored:
            explored.add(node)

            for child,weight in graph[node].items():
                total_cost = cost + weight
                frontier.append((child, path + [child],total_cost))

    print("Goal not found")                


bfs_with_cost(graph2, 'S', 'G')
```

## DFS TRAVERSAL

```python
# Graph (same style)
graph = {
    'S': ['A', 'C', 'D'],
    'A': [],
    'B': ['E'],
    'C': ['F'],
    'D': ['B', 'G'],
    'E': ['G'],
    'F': ['E', 'G'],
    'G': []
}

def dfs(graph, start):
    explored = set()
    frontier = [start]

    print("DFS Traversal:", end=" ")

    while frontier:
        node = frontier.pop()   # LIFO

        if node not in explored:
            print(node, end=" ")
            explored.add(node)

            # reverse order for correct traversal
            for child in reversed(graph[node]):
                frontier.append(child)

dfs(graph, 'S')
```

## UCS (Uniform Cost Search)

```python
import heapq

graph = {
    'S': {'A':3,'C':2,'D':2},
    'A': {},
    'B': {'E':2},
    'F': {'E':0,'G':4},
    'D': {'B':3,'G':8},
    'G': {},
    'E': {'G':2},
    'C': {'F':1}
}

graph2 = {
    'S':{'D':3,'E':9,'P':1},
    'D':{'B':1,'C':8, 'E': 2},
    'E':{'R':1,'H':8},
    'P':{'Q':15},
    'B':{'A':2},
    'C':{'A':3},
    'R':{'F':1},
    'H':{'P':2,'Q':1},
    'F':{'G':2,'C':1},
    'P':{'Q':15},
    'Q':{},
    'A':{},
    'G':{}

}
def ucs(graph, start, goal):
    explored = set()
    
    # priority queue: (cost, node, path)
    frontier = [(0, start, [start])]

    print("UCS Traversal:\n")

    while frontier:
        cost, node, path = heapq.heappop(frontier)

        if node in explored:
            continue

        explored.add(node)

        print(f"Visited: {node} with cost {cost}")

        if node == goal:
            print("\nGoal Found!")
            print("Path:", path)
            print("Total Cost:", cost)
            return

        for child, weight in graph[node].items():
            if child not in explored:
                heapq.heappush(frontier, (cost + weight, child, path + [child]))

    print("Goal not found")

# Run
ucs(graph2, 'S', 'G')
```

# DFS

```python
graph = {
    'S': ['A', 'B'],
    'A': ['C'],
    'B': [],
    'C': ['G'],
    'G': []
}


def DLS(graph, current, goal, limit, depth=0, path=None):
    
    if path is None:
        path = []

    # Add current node to path
    path.append(current)

    # Goal Test
    if current == goal:
        return path

    # Depth limit reached → cutoff
    if depth == limit:
        return None

    # Explore children (DFS style)
    for neighbor in graph[current]:
        if neighbor not in path:   # avoid cycles
            result = DLS(graph, neighbor, goal, limit, depth+1, path.copy())
            
            if result is not None:
                return result

    return None
result = DLS(graph, 'S', 'G', limit=3)
print("Result:", result)
```

# IDDFS

```python
graph = {
    'S': ['A', 'B'],
    'A': ['C'],
    'B': [],
    'C': ['G'],
    'G': []
}

def DLS(graph, current, goal, limit, depth=0, path=None):
    
    if path is None:
        path = []

    path.append(current)

    if current == goal:
        return path

    if depth == limit:
        return None

    for neighbor in graph[current]:
        if neighbor not in path:
            result = DLS(graph, neighbor, goal, limit, depth+1, path.copy())
            
            if result is not None:
                return result

    return None


def IDDFS(graph, start, goal, max_depth):
    
    for limit in range(max_depth + 1):
        print(f"\nTrying depth limit = {limit}")
        
        result = DLS(graph, start, goal, limit)
        
        if result is not None:
            return result
    
    return None

result = IDDFS(graph, 'S', 'G', max_depth=5)
print("\nFinal Path:", result)
```
