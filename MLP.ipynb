# BFS

```python
from collections import deque

graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'F':{'E':0,'G':4},
    'D':{'B':3,'G':8},
    'G':{},
    'E':{'G':2},
    'C':{'F':1}

}

def bfs(graph, start, goal):
    explored = set()
    frontier = deque([(start,[start],0)])

    while frontier:
        node,path,cost = frontier.popleft()

        if node == goal:
            print("GOAL FOUND !! \nPATH = ",path,"\nCOST = $",cost)
            return 

        if node not in explored:
            explored.add(node)
        
        for child,weights in graph[node].items():
            total_cost = cost+weights
            frontier.append((child, path+[child],total_cost))
    
    print("PATH NOT FOUND!!!")


bfs(graph, 'S', 'G')
```

# DFS

```python
from collections import deque

graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'F':{'E':0,'G':4},
    'D':{'B':3,'G':8},
    'G':{},
    'E':{'G':2},
    'C':{'F':1}

}

def dfs(graph, start, goal):
    explored = set()
    frontier = deque([(start,[start],0)])

    while frontier:
        node,path,cost = frontier.pop()

        if node == goal:
            print("GOAL FOUND !! \nPATH = ",path,"\nCOST = $",cost)
            return 

        if node not in explored:
            explored.add(node)
        
        for child,weights in reversed(list(graph[node].items())):
            total_cost = cost+weights
            frontier.append((child, path+[child],total_cost))
    
    print("PATH NOT FOUND!!!")


dfs(graph, 'S', 'G')

```

# UCS

```python
import heapq 
graph2 = {
    'S':{'D':3,'E':9,'P':1},
    'D':{'B':1,'C':8, 'E': 2},
    'E':{'R':1,'H':8},
    'P':{'Q':15},
    'B':{'A':2},
    'C':{'A':3},
    'R':{'F':1},
    'H':{'P':2,'Q':1},
    'F':{'G':2,'C':1},
    'P':{'Q':15},
    'Q':{},
    'A':{},
    'G':{}

}

def UCS(graph,start,goal):

    explored = set()
    frontier = [(0, start, [start])]

    while frontier:
        cost,node,path = heapq.heappop(frontier)

        if node in explored:
            continue

        explored.add(node)
      

        print("Visited: ",node, "cost $",cost)

        if node == goal:
            print("\nGoal found!!!\nPath = ", path, "\nCost $",cost)
            return

        for child, weights in graph[node].items():
            if child not in explored:
                heapq.heappush(frontier, (cost+weights, child, path+[child]))

        
    print("GOAL NOT FOUND!!")

    
    
    
UCS(graph2, 'S', 'G')
print("HEllo")
```

# DLS (Depth Limit Search)

```python
graph = {
    'A': {'B': 2, 'C': 3},
    'B': {'D': 4, 'E': 1},
    'C': {'F': 5},
    'D': {},
    'E': {},
    'F': {}
}

def DLS(graph, node, goal, depth, path, cost):
    print("Visited: ", node, " depth: ", depth, " cost $", cost)

    if node == goal:
        print("Goal Found: ",path, "depth : ",depth, " cost $", cost)
        return True

    if depth == 0:
        return False
    
    for child,weights in graph[node].items():
        found = DLS(graph,child,goal,depth-1,path+[child],cost+weights)
        if found:
            return True
    return False

DLS(graph,'A','E',3,['A'],0)
```

# IDDFS

```python
graph = {
    'A': {'B': 2, 'C': 3},
    'B': {'D': 4, 'E': 1},
    'C': {'F': 5},
    'D': {},
    'E': {},
    'F': {}
}

def DLS(graph, node, goal, depth, path, cost):

    if node == goal:
        return path, cost

    if depth == 0:
        return None

    for child, weight in graph[node].items():
        result = DLS(graph,child,goal,depth - 1,path + [child],cost + weight)

        if result is not None:
            return result

    return None

def IDDFS(graph, start, goal, max_depth):

    for depth in range(max_depth + 1):
        print(f"\n--- Trying depth {depth} ---")

        result = DLS(graph, start, goal, depth, [start], 0)

        if result is not None:
            print("\nGoal Found!")
            print("Path:", result[0])
            print("Cost:", result[1])
            return

    print("Goal not found")


 

IDDFS(graph, 'A', 'E', 3)
```

# GBFS (Greedy Best First Search)
## Uses UCS

```python
import heapq

graph = {
    'S': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['E'],
    'C': [],
    'D': [],
    'E': []
}

heuristic = {
    'S': 7,
    'A': 6,
    'B': 2,
    'C': 1,
    'D': 3,
    'E': 0   # goal
}

def GBFS(graph, start, goal, heuristic):

    explored = set()
    frontier = [(heuristic[start], start, [start])]

    while frontier:

        h, node, path = heapq.heappop(frontier)

        if node in explored:
            continue

        explored.add(node)

        print("Visited:", node, "h(n):", h)

        if node == goal:
            print("\nGoal Found!")
            print("Path:", path)
            return

        for child in graph[node]:
            if child not in explored:
                heapq.heappush(frontier,(heuristic[child], child, path + [child]))

    print("Goal not found")

    
GBFS(graph, 'S', 'E', heuristic)
```

# A* Algorithm

```python
graph = {
    'S': {'A': 1, 'B': 4},
    'A': {'C': 2, 'D': 5},
    'B': {'D': 1},
    'C': {'G': 5},
    'D': {'G': 1},
    'G': {}
}

heuristic = {
    'S': 7,
    'A': 6,
    'B': 2,
    'C': 3,
    'D': 1,
    'G': 0
}

import heapq

def A_star(graph, start, goal, heuristic):

    explored = set()

    # (f, g, node, path)
    frontier = [(heuristic[start], 0, start, [start])]

    while frontier:

        f, g, node, path = heapq.heappop(frontier)

        if node in explored:
            continue

        explored.add(node)

        print("Visited:", node, "g(n):", g, "h(n):", heuristic[node], "f(n):", f)

        if node == goal:
            print("\nGoal Found!")
            print("Path:", path)
            print("Cost:", g)
            return path, g

        for child, weight in graph[node].items():

            if child not in explored:

                new_g = g + weight
                new_f = new_g + heuristic[child]

                heapq.heappush(frontier,(new_f, new_g, child, path + [child]))

    print("Goal not found")
    return None

A_star(graph, 'S', 'G', heuristic)
```

# HILL CLimb

```python
def hill_climbing(graph, start, heuristic):

    current = start
    path = [current]

    while True:

        neighbors = graph[current]

        if not neighbors:
            break

        next_node = None
        best_value = heuristic[current]

        for n in neighbors:
            if heuristic[n] < best_value:
                best_value = heuristic[n]
                next_node = n

        if next_node is None:
            break

        current = next_node
        path.append(current)

    return path
```
