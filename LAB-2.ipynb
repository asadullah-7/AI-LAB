```python
def mean_median_mode(data):
    n = len(data)

    # Mean
    mean = sum(data) / n

    # Median
    sorted_data = sorted(data)
    if n % 2 == 1:
        median = sorted_data[n // 2]
    else:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2

    # Mode
    freq = {}
    for num in data:
        freq[num] = freq.get(num, 0) + 1

    max_freq = max(freq.values())
    mode = [k for k, v in freq.items() if v == max_freq]

    # if only one mode return value otherwise list
    if len(mode) == 1:
        mode = mode[0]

    return mean, median, mode


# Example
data = [1, 2, 2, 3, 4, 5]

mean, median, mode = mean_median_mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
import time
del list

my_big_list = list(range(100000, 300001))

def search(integer_list, key):
    for i in range(len(integer_list)):
        if integer_list[i] == key:
            return i
    return None

def timer(key):
    start = time.time()
    search(my_big_list, key)
    end = time.time()
    return end - start

print("Front element (100000):", timer(100000))
print("Middle element (200000):", timer(200000))
print("Back element (300000):", timer(300000))
print("Not in list (3):", timer(3))
```
