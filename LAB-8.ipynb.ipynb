```python
import random
import math

# Board setup
board = [' ' for _ in range(9)]

# Print board
def print_board():
    print()
    for i in range(3):
        print(board[i*3] + " | " + board[i*3+1] + " | " + board[i*3+2])
        if i < 2:
            print("--+---+--")
    print()

# Check winner
def check_winner(b, player):
    win_states = [
        [0,1,2], [3,4,5], [6,7,8],  # rows
        [0,3,6], [1,4,7], [2,5,8],  # cols
        [0,4,8], [2,4,6]            # diagonals
    ]
    for state in win_states:
        if all(b[i] == player for i in state):
            return True
    return False

# Check draw
def is_draw(b):
    return ' ' not in b

# Get available moves
def get_moves(b):
    return [i for i in range(9) if b[i] == ' ']

# Heuristic function (Exercise 8.4)
def heuristic(b):
    score = 0
    lines = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    
    for line in lines:
        x_count = sum(1 for i in line if b[i] == 'X')
        o_count = sum(1 for i in line if b[i] == 'O')
        
        if x_count > 0 and o_count == 0:
            score += x_count
        elif o_count > 0 and x_count == 0:
            score -= o_count
    
    return score

# Minimax with depth limit (bounded lookahead)
def minimax(b, depth, is_max, max_depth=3):
    
    if check_winner(b, 'X'):
        return 10 - depth
    if check_winner(b, 'O'):
        return depth - 10
    if is_draw(b):
        return 0
    
    # Cutoff (bounded lookahead)
    if depth == max_depth:
        return heuristic(b)
    
    moves = get_moves(b)
    
    if is_max:
        best = -math.inf
        for move in moves:
            b[move] = 'X'
            val = minimax(b, depth+1, False, max_depth)
            b[move] = ' '
            best = max(best, val)
        return best
    else:
        best = math.inf
        for move in moves:
            b[move] = 'O'
            val = minimax(b, depth+1, True, max_depth)
            b[move] = ' '
            best = min(best, val)
        return best

# Get best move with random tie-break
def best_move():
    moves = get_moves(board)
    best_score = -math.inf
    best_moves = []
    
    for move in moves:
        board[move] = 'X'
        score = minimax(board, 0, False)
        board[move] = ' '
        
        if score > best_score:
            best_score = score
            best_moves = [move]
        elif score == best_score:
            best_moves.append(move)
    
    return random.choice(best_moves)

# Human move
def human_move():
    moves = get_moves(board)
    move = random.choice(moves)  # Random (as required)
    board[move] = 'O'
    print(f"Human chose position {move}")

# Game loop
def play_game():
    print("Tic Tac Toe (Computer = X, Human = O)")
    print_board()
    
    while True:
        # Computer move
        move = best_move()
        board[move] = 'X'
        print(f"Computer chose position {move}")
        print_board()
        
        if check_winner(board, 'X'):
            print("Computer wins!")
            break
        if is_draw(board):
            print("Draw!")
            break
        
        # Human move
        human_move()
        print_board()
        
        if check_winner(board, 'O'):
            print("Human wins!")
            break
        if is_draw(board):
            print("Draw!")
            break

# Run game
play_game()
```
