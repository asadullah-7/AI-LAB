```python
import heapq

graph = {
    'S': {'A': 3, 'C': 2, 'D': 2},
    'A': {},
    'B': {'E': 2},
    'F': {'E': 0, 'G': 4},
    'D': {'B': 3, 'G': 8},
    'G': {},
    'E': {'G': 2},
    'C': {'F': 1}
}

def ucs(graph, start, goal):
    visited = set()
    
    # Priority queue: (cost, node, path)
    pq = [(0, start, [start])]
    
    explored_order = []
    
    while pq:
        cost, current, path = heapq.heappop(pq)
        
        if current in visited:
            continue
        
        visited.add(current)
        explored_order.append(current)
        
        if current == goal:
            return path, cost, explored_order
        
        # Sort neighbors alphabetically (tie-breaking rule)
        for neighbor in sorted(graph[current]):
            new_cost = cost + graph[current][neighbor]
            heapq.heappush(pq, (new_cost, neighbor, path + [neighbor]))
    
    return None, float("inf"), explored_order


path, cost, explored = ucs(graph, 'S', 'G')

print("Explored Order:", explored)
print("Path:", path)
print("Total Cost:", cost)
```

```python
graph = {
    'S': {'A': 3, 'C': 2, 'D': 2},
    'A': {},
    'B': {'E': 2},
    'F': {'E': 0, 'G': 4},
    'D': {'B': 3, 'G': 8},
    'G': {},
    'E': {'G': 2},
    'C': {'F': 1}
}

def dls(current, goal, limit, visited, path, cost, depth, explored):
    
    explored.append(current)
    
    if current == goal:
        return path, cost
    
    if depth == limit:
        return None, float("inf")
    
    visited.add(current)
    
    for neighbor, weight in sorted(graph[current].items()):
        if neighbor not in visited:
            result_path, total_cost = dls(
                neighbor, goal, limit,
                visited.copy(),
                path + [neighbor],
                cost + weight,
                depth + 1,
                explored
            )
            
            if result_path:
                return result_path, total_cost
    
    return None, float("inf")


explored = []
path, cost = dls('S', 'G', 2, set(), ['S'], 0, 0, explored)

print("Explored Order:", explored)
print("Path:", path)
print("Cost:", cost)
```

YES, it finds the goal

Reason:
Depth limit = 2

Goal node G is reachable at depth = 2 via:

S -> D -> G
Since this is within the limit, DLS successfully reaches G
