```python
initial_state = [
    [2, 1, 3],
    [4, 7, 6],
    [5, 8, 0]
]

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
```

```python
def goal_test(state):
    return state == goal_state
```

```python
def get_neighbors(state):
    neighbors = []
    
    # find blank position
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                x, y = i, j

    moves = [(-1,0),(1,0),(0,-1),(0,1)]  # up, down, left, right
    
    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state] # Yeh deep copy kry ga......
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    
    return neighbors
```

```python
def h1(state):
    count = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != goal_state[i][j]:
                count += 1
    return count
```

```python
def h2(state):
    distance = 0
    
    for i in range(3):
        for j in range(3):
            value = state[i][j]
            
            if value != 0:
                goal_x = (value - 1) // 3
                goal_y = (value - 1) % 3
                
                distance += abs(i - goal_x) + abs(j - goal_y)
    
    return distance
```

```python
import heapq

def GreedySearch(initial_state, heuristic):
    visited = set()
    pq = []
    
    # (heuristic, state, path)
    heapq.heappush(pq, (heuristic(initial_state), initial_state, []))
    
    while pq:
        h, state, path = heapq.heappop(pq)
        
        if goal_test(state):
            return path + [state]
        
        visited.add(str(state))
        
        for neighbor in get_neighbors(state):
            if str(neighbor) not in visited:
                heapq.heappush(pq, (heuristic(neighbor), neighbor, path + [state]))
    
    return None
```

```python
solution = GreedySearch(initial_state, h1)

for step in solution:
    for row in step:
        print(row)
    print("------")
```

```python
solution = GreedySearch(initial_state, h2)

for step in solution:
    for row in step:
        print(row)
    print("------")
```
