```python
graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'F':{'E':0,'G':4},
    'D':{'B':3,'G':8},
    'G':{},
    'E':{'G':2},
    'C':{'F':1}

}

graph2={
    'A':{'B':11,'C':14,'D':7},
    'B':{'E':15},
    'C':{'E':8,'F':10},
    'D':{'F':25},
    'E':{'H':9},
    'F':{'G':20},
    'H':{'G':10},

}

def bfs(graph,start,goal):
  visited=set()
  queue=[(start,[start],0)]
  while queue:
    current,path,cost=queue.pop(0)

    if current==goal:
      return path,cost

    if current not in visited:
      visited.add(current)
      for neighbor,weight in graph[current].items():
        total_cost=cost+weight
        queue.append((neighbor,path +[neighbor],total_cost))
  return None,float("inf")

bfs_path,bfs_cost=bfs(graph,'S','G')
print(f"Path:{bfs_path} and Cost:{bfs_cost}")

bfs_path,bfs_cost=bfs(graph2,'A','G')
print(f"Path:{bfs_path} and Cost:{bfs_cost}")


```

```python
graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'F':{'E':0,'G':4},
    'D':{'B':3,'G':8},
    'G':{},
    'E':{'G':2},
    'C':{'F':1}

}

def dfs(current,goal,visited,path,cost):
  if current == goal:
    return path,cost

  visited.add(current)

  for neighbor,weight in sorted(graph[current].items()):

    if neighbor not in visited:
      result_path,total_cost=dfs(neighbor,goal,visited.copy(), path + [neighbor], cost + weight)

      if result_path: # If a path to the goal was found
                return result_path, total_cost

  return None, float("inf")

# Execution
path, cost = dfs('S', 'G', set(), ['S'], 0)
print(f"Path: {path} and Cost: {cost}")





```
